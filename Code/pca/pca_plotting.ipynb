```python
import numpy as np
from datetime import date
import matplotlib
import matplotlib.cm as cmx
from matplotlib import pyplot as plt
```

```python
npzfile1 = np.load('/gpfs/alpine/gen150/scratch/arjun2612/ORNL_Coding/Code/pca/smc2_plotting.npz')
npzfile2 = np.load('/gpfs/alpine/gen150/scratch/arjun2612/ORNL_Coding/Code/pca/smc2_sk_clusterfiles.npz')

smc2_result = npzfile1['res']
smc2_label_validation = npzfile1['labval']
smc2_reduced_val = npzfile2['redval']
```

```python
npzfile3 = np.load('/gpfs/alpine/gen150/scratch/arjun2612/ORNL_Coding/Code/pca/hea_plotting.npz')
npzfile4 = np.load('/gpfs/alpine/gen150/scratch/arjun2612/ORNL_Coding/Code/pca/hea_sk_clusterfiles.npz')

hea_result = npzfile3['res']
hea_label_validation = npzfile3['labval']
hea_reduced_val = npzfile4['redval']
```

```python
today = date.today()
d = today.strftime("%b-%d-%Y")
```

```python
# Plotting for the smc2 dataset
cNorm = matplotlib.colors.Normalize(vmin=min(smc2_result), vmax=max(smc2_result))
scalarMap = cmx.ScalarMappable(norm=cNorm)
fig = plt.figure(1)
plt.scatter(smc2_reduced_val[:, 0], smc2_reduced_val[:, 1], c=scalarMap.to_rgba(smc2_result))
scalarMap.set_array(smc2_result)
cbar = fig.colorbar(scalarMap)
cbar.set_ticks([0,1,2])
cbar.set_ticklabels(["COV2", "MERS", "SARS"])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Cluster Map of Predicted Set')
# fig.savefig('pca_pred_' + d + '.png')
plt.show()
```

```python
cNorm = matplotlib.colors.Normalize(vmin=min(smc2_label_validation), vmax=max(smc2_label_validation))
scalarMap = cmx.ScalarMappable(norm=cNorm)
fig = plt.figure(2)
plt.scatter(smc2_reduced_val[:, 0], smc2_reduced_val[:, 1], c=scalarMap.to_rgba(smc2_label_validation))
scalarMap.set_array(smc2_label_validation)
cbar = fig.colorbar(scalarMap)
cbar.set_ticks([0,1,2])
cbar.set_ticklabels(["COV2", "MERS", "SARS"])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Cluster Map of Validation Set')
# fig.savefig('pca_val_' + d + '.png')
plt.show()
```

```python
# Plotting for the HEA dataset
cNorm = matplotlib.colors.Normalize(vmin=min(hea_result), vmax=max(hea_result))
scalarMap = cmx.ScalarMappable(norm=cNorm)
fig = plt.figure(1)
plt.scatter(hea_reduced_val[:, 0], hea_reduced_val[:, 1], c=scalarMap.to_rgba(hea_result))
scalarMap.set_array(hea_result)
cbar = fig.colorbar(scalarMap)
cbar.set_ticks([0,1,2,3,4])
cbar.set_ticklabels(["50", "83", "200", "524", "1500"])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Cluster Map of Predicted Set')
# fig.savefig('pca_pred_' + d + '.png')
plt.show()
```
