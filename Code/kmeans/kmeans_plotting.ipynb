```python
import matplotlib
import matplotlib.cm as cmx
from matplotlib import pyplot as plt
import numpy as np
from datetime import date
```

```python
today = date.today()
d = today.strftime("%b-%d-%Y")
```

```python
npzfile1 = np.load('/gpfs/alpine/gen150/scratch/arjun2612/ORNL_Coding/Code/kmeans/smc2_sk_plotting.npz')

smc2_labval = npzfile1['reslab']
smc2_label_validation = npzfile1['lv']
smc2_reduced_val = npzfile1['redval']
```

```python
npzfile2 = np.load('/gpfs/alpine/gen150/scratch/arjun2612/ORNL_Coding/Code/kmeans/hea_sk_plotting.npz')

hea_labval = npzfile2['reslab']
hea_label_validation = npzfile2['lv']
hea_reduced_val = npzfile2['redval']
```

```python
# Plotting for smc2 dataset
cNorm = matplotlib.colors.Normalize(vmin=min(smc2_labval), vmax=max(smc2_labval))
scalarMap = cmx.ScalarMappable(norm=cNorm)
fig = plt.figure(5)
plt.scatter(smc2_reduced_val[:, 0], smc2_reduced_val[:, 1], c=scalarMap.to_rgba(smc2_labval))
scalarMap.set_array(smc2_labval)
cbar = fig.colorbar(scalarMap)
cbar.set_ticks([0,1,2])
cbar.set_ticklabels(["COV2", "MERS", "SARS"])
plt.title('K-Means Cluster Map of Predicted Set')
# fig.savefig('kmeans_pred_' + d + '.png')
plt.show()
```

```python
cNorm = matplotlib.colors.Normalize(vmin=min(smc2_label_validation), vmax=max(smc2_label_validation))
scalarMap = cmx.ScalarMappable(norm=cNorm)
fig = plt.figure(6)
plt.scatter(smc2_reduced_val[:, 0], smc2_reduced_val[:, 1], c=scalarMap.to_rgba(smc2_label_validation))
scalarMap.set_array(smc2_label_validation)
cbar = fig.colorbar(scalarMap)
cbar.set_ticks([0,1,2])
cbar.set_ticklabels(["COV2", "MERS", "SARS"])
plt.title('K-Means Cluster Map of Validation Set')
# fig.savefig('kmeans_val_' + d + '.png')
plt.show()
```

```python
# Plotting for HEA dataset
cNorm = matplotlib.colors.Normalize(vmin=min(hea_labval), vmax=max(hea_labval))
scalarMap = cmx.ScalarMappable(norm=cNorm)
fig = plt.figure(5)
plt.scatter(hea_reduced_val[:, 0], hea_reduced_val[:, 1], c=scalarMap.to_rgba(hea_labval))
scalarMap.set_array(hea_labval)
cbar = fig.colorbar(scalarMap)
cbar.set_ticks([0,1,2,3,4])
cbar.set_ticklabels(["50", "83", "200", "524", "1500"])
plt.title('K-Means Cluster Map of Predicted Set')
# fig.savefig('kmeans_pred_' + d + '.png')
plt.show()
```

```python
cNorm = matplotlib.colors.Normalize(vmin=min(hea_label_validation), vmax=max(hea_label_validation))
scalarMap = cmx.ScalarMappable(norm=cNorm)
fig = plt.figure(6)
plt.scatter(hea_reduced_val[:, 0], hea_reduced_val[:, 1], c=scalarMap.to_rgba(hea_label_validation))
scalarMap.set_array(hea_label_validation)
cbar = fig.colorbar(scalarMap)
cbar.set_ticks([0,1,2,3,4])
cbar.set_ticklabels(["50", "83", "200", "524", "1500"])
plt.title('K-Means Cluster Map of Validation Set')
# fig.savefig('kmeans_val_' + d + '.png')
plt.show()
```
